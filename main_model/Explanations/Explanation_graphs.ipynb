```python
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
f1 = lambda x: 10 + 0.2*x
f2 = lambda x: 5 + 0.2*x

v1 = lambda x: 7 + 0.2*x 
v2 = lambda x: 5.0 + 0.4*x

xs = np.arange(0,13,4)
x2 = np.arange(7,20,4)
```

```python
fig,ax = plt.subplots(1,1)


ax.plot(xs,f1(xs),color="blue",marker='.')
ax.plot(x2,f2(x2),color="blue",marker='.')

for i,v in enumerate(np.append(xs,x2)):
    if i < len(xs):
        ax.text(v,f1(v)-0.7,f'{i}')
    else: 
        ax.text(v,f2(v)-0.7,f'{i}')

ax.plot([7,12],[6.4,12.4],color="grey",linestyle="dashed")

ax.set_ylim([0,15])

ax.vlines(12.4,0,15,color="red",label="$\\bar{a}$")
ax.set_xlabel("M")
ax.set_ylabel("C")
ax.legend()
```

```python
fig,ax = plt.subplots(2,2,figsize=(8,7))


ax[0,0].plot(xs,v1(xs),color="blue",marker='.')
ax[0,0].plot(x2,v2(x2),color="blue",marker='.')


ax[0,1].plot(xs,f1(xs),color="blue",marker='.')
ax[0,1].plot(x2,f2(x2),color="blue",marker='.')

for i,v in enumerate(np.append(xs,x2)):
    if i < len(xs): 
        ax[0,0].text(v,v1(v)-0.7,f'{i}')
        ax[0,1].text(v,f1(v)-0.7,f'{i}')
    else: 
        ax[0,0].text(v,v2(v)-0.5,f'{i}') 
        ax[0,1].text(v,f2(v)-0.7,f'{i}')

ax[0,1].plot([7,12],[6.4,12.4],color="grey",linestyle="dashed")
ax[0,1].plot([7,12],[6.4,12.4],color="grey",linestyle="dashed")

ax[0,1].arrow(x2[0],f2(x2[0]),0,f1(x2[0])- f2(x2[0]) - 0.2 ,length_includes_head=True, head_width =0.3 )
ax[0,1].arrow(xs[-1],f1(xs[-1]),0,f2(xs[-1])- f1(xs[-1]) + 0.2,length_includes_head=True, head_width =0.3 )

ax[0,0].set_ylim([0,15])
ax[0,1].set_ylim([0,15])
ax[0,0].vlines(12.4,0,15,color="red",label="$\\bar{a}$")
ax[0,0].vlines(10,0,15,color="cyan",label="$\\bar{a}_2$")
ax[0,1].vlines(12.4,0,15,color="red",label="$\\bar{a}$")
ax[0,1].vlines(10,0,15,color="cyan",label="$\\bar{a}_2$")
ax[0,0].set_xlabel("M")
ax[0,0].set_ylabel("V")
ax[0,1].set_xlabel("M")
ax[0,1].set_ylabel("C")
ax[0,0].legend()


ax[1,0].plot([xs[0],10],[v1(xs[0]),v1(10)],color="blue")
ax[1,0].plot([10,x2[-1]],[v2(10),v2(x2[-1])],color="blue")
ax[1,0].vlines(12.4,0,15,color="red",label="$\\bar{a}$")
ax[1,0].vlines(10,0,15,color="cyan",label="$\\bar{a}_2$")
ax[1,0].set_ylim([0,15])
ax[1,1].set_ylim([0,15])

ax[1,1].plot([xs[0],10],[f1(xs[0]),f1(10)],color="blue")
ax[1,1].plot([10,x2[-1]],[f2(10),f2(x2[-1])],color="blue")
ax[1,1].vlines(12.4,0,15,color="red",label="$\\bar{a}$")
ax[1,1].vlines(10,0,15,color="cyan",label="$\\bar{a}_2$")
ax[1,1].set_ylim([0,15])
ax[1,1].set_ylim([0,15])
ax[1,0].set_xlabel("M")
ax[1,0].set_ylabel("V")
ax[1,1].set_xlabel("M")
ax[1,1].set_ylabel("C")

ax[0,0].set_title("Choice Specific Value Function")
ax[0,1].set_title("Consumption Correspondance")

ax[1,0].set_title("Refined Choice Specific Value Function")
ax[1,1].set_title("Refined Consumption function")
fig.tight_layout()
```
